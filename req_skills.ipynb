```python
import pandas as pd
import math
```

```python
df = pd.read_csv('resume_data.csv')
df
```

```python
df.columns
```

```python
len(df['job_position_name'].unique())
```

```python
temp_df = df[['skills_required', 'job_position_name', 'educationaL_requirements', 'experiencere_requirement']]
temp_df
```

```python
job_summary = (
    temp_df.groupby('job_position_name')
    .agg({
        'skills_required': lambda x: list(set(sum((i.split(',') for i in x if isinstance(i, str)), []))),
        'educationaL_requirements': lambda x: list(set(x)),
        'experiencere_requirement': lambda x: list(set(x))
    })
    .reset_index()
)
```

```python
job_knowledge = {
    row['job_position_name']: {
        "skills": row['skills_required'],
        "education": row['educationaL_requirements'],
        "experience": row['experiencere_requirement']
    }
    for _, row in job_summary.iterrows()
}
```

```python
def clean_text_list(lst):
    if not lst or all((isinstance(x, float) and math.isnan(x)) for x in lst):
        return "Not specified"
    return ', '.join(
        i.replace('\n', ', ').replace('•', '').strip() 
        for i in lst if isinstance(i, str) and i.strip()
    )
```

```python
text = []

for idx, (job, info) in enumerate(job_knowledge.items()):
    skills = clean_text_list(info.get("skills", []))
    education = clean_text_list(info.get("education", []))
    experience = clean_text_list(info.get("experience", []))

    cleaned_output = ""
    
    cleaned_output += f"{idx+1} -> {job}\n"
    cleaned_output += f"Skills: {skills}\n"
    cleaned_output += f"Education: {education}\n"
    cleaned_output += f"Experience: {experience}\n\n"

    text.append(cleaned_output)
```

```python
file_path = 'all_jobs_req_skills_exp_edu.txt'

with open(file_path, 'w') as f:
    f.write('\n'.join(text))
```
